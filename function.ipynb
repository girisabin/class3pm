```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def cal():
    a=float(input("enter first number"))
    b=float(input("enter second number"))
    o=str(input("Enter operator + - * / "))
    if o=='+':
        add(a,b)
       

    elif o=='-':
        sub(a,b)

    elif o=='*':
          mul(a,b)
    elif o=='/' and b!=0:
        div(a,b)
        
    elif o=='/' and b==0:
        print("Value of b cannot be zero")

    else:
        print("You enter invalid operator !!!!!!")
        
cal()
```
